```python
from fitter import Fitter, get_common_distributions, get_distributions
```

```python
Potential_Gross = df["Potential_Gross"].values
f = Fitter(Potential_Gross,
           distributions=['gamma',
                          'lognorm',
                          'pareto','lognorm','uniform','exponpow'])
f.fit()
f.summary()
```

```python
f.get_best(method = 'sumsquare_error')
```

```python
Seats_Sold = df["Seats_Sold"].values
g = Fitter(Seats_Sold,
           distributions=['gamma',
                          'lognorm',
                          "beta",
                          "expon",
                          "norm"])
g.fit()
g.summary()
```

```python
g.get_best(method = 'sumsquare_error')
```

```python
Average_ticket = df["Average_ticket"].values
h = Fitter(Average_ticket,
           distributions=['gamma',
                          'lognorm',
                          "beta",
                          "expon",
                          "norm"])
h.fit()
h.summary()
```
